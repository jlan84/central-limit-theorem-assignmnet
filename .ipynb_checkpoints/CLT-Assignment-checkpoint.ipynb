```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_draws(dist, params, size=200):
    """
    Draw a sample from a specified distribution
    with given parameters and return these in an array.

    Parameters
    ----------

    dist: scipy.stats distribution object:
      Distribution object from scipy.stats, must have a .rvs method

    params: dict:
      Parameters needed to define the distribution dist.
    For example, if dist = scipy.stats.binom, then params could be

          {'n': 100, 'p': 0.25}

    size: int:
      The number of values to draw

    Returns
    -------
    sample: np.array, shape (size, )
      An i.i.d sample from the specified distribution.
    """
    return dist(**params).rvs(size)
```

```python
def plot_means(dist, params, ax, size=200, repeat=10):
    mean_list = []
    for i in range(repeat):
        mean_list.append(np.mean(make_draws(dist=dist, params=params, size=size)))
    ax.hist(mean_list, density=True, bins=30)
    plt.show()
    
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.binom, {'n':1000, 'p': 0.5}, ax, repeat=1000)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.poisson, {'mu':1}, ax, repeat=1000)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.expon, {'scale':1}, ax, repeat=1000)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.geom, {'p':.5}, ax, repeat=1000)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.uniform, {'loc': 0, 'scale':1}, ax, size=500, repeat=5000)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.uniform, {'loc': 0, 'scale':1}, ax, size=10, repeat=5000)
```

The standard deviation for the sample set increases when decreasing the size

Because the standard deviation is related to the sample size (sigma/sqrt(sample size)) and not the amount of times the samples were drawn <br />\
the distribution will become more normal with an increase in repeat sampling but the std dev will not get smaller.


```python
def plot_means(dist, params, ax, repeat, size=200):
  meansList = []
  for i in range(repeat):
    meansList.append(np.mean(make_draws(dist, params, size=size)))
  ax.hist(meansList, bins=50)

sizeList = [100, 1000, 10000, 20000]

fig, axs = plt.subplots(2,2, figsize=(12,12))
for size, ax in zip(sizeList, axs.flatten()):
  plot_means(stats.binom, {'n':30, 'p': 0.5}, ax, repeat=size)


plt.show()

```

```python
sizeList = [100, 1000, 10000, 20000]

fig, axs = plt.subplots(2,2, figsize=(12,12))
for size, ax in zip(sizeList, axs.flatten()):
  plot_means(stats.binom, {'n':30, 'p': 0.5}, ax, repeat=size, size=10)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
plot_means(stats.uniform, {'loc': 0, 'scale':1}, ax, size=10, repeat=5000)
```

```python
def plot_max(dist, params, ax, repeat, size=200):
  maxList = []
  for i in range(repeat):
    maxList.append(np.max(make_draws(dist, params, size=size)))
  ax.hist(maxList, bins=50)

sizeList = [100, 1000, 10000, 20000]

fig, axs = plt.subplots(2,2, figsize=(12,12))
for size, ax in zip(sizeList, axs.flatten()):
  plot_max(stats.binom, {'n':30, 'p': 0.5}, ax, repeat=size)
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
for size, ax in zip(sizeList, axs.flatten()):
  plot_max(stats.poisson, {'mu':30}, ax, repeat=size)
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,12))
for size, ax in zip(sizeList, axs.flatten()):
  plot_max(stats.uniform, {'loc':10, 'scale': 100}, ax, repeat=size)
plt.show()
```
